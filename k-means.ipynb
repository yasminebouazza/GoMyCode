```python
import pandas as pd
df = pd.read_csv('C:/Users/yessm/Downloads/CC GENERAL.csv')
df.head(20)
```

```python
df = df.drop(['CUST_ID'], axis=1) 
df
```

```python
df.isnull().sum()
```

```python
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mean() ,inplace=True)
df 
```

```python
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].mean() ,inplace=True)
df 
```

```python
df.isnull().sum().sum()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward')
clust_labels1=model.fit_predict(df)
agglomerative = pd.DataFrame(clust_labels1)
agglomerative
```

```python
import matplotlib.pyplot as plt 
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(df.BALANCE,df.PAYMENTS,c=agglomerative[0],s=50)
plt.colorbar(scatter);
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(df.BALANCE,df.MINIMUM_PAYMENTS,c=agglomerative[0],s=50)
plt.colorbar(scatter);
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(df.BALANCE,df.CASH_ADVANCE,c=agglomerative[0],s=50)
plt.colorbar(scatter);
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters = 5, random_state=0)
kmeans.fit(df)
labels = pd.DataFrame(kmeans.labels_)
labels
kmeans.predict(df)
print(kmeans.cluster_centers_)
centroids=kmeans.cluster_centers_
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize =(25,7))
dend =shc.dendrogram(shc.linkage(df,method='complete'))
```

```python
sse=[]
kPlot=[]
for k in range(1,15):
    kmeans=KMeans(n_clusters = k, random_state=0)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)
    kPlot.append(k)
    
plt.plot(kPlot,sse);
```

```python
kmeans=KMeans(n_clusters = 4, random_state=0)
kmeans.fit(df)
kmeans.predict(df)
```
