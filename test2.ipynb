```python
import pandas as pd
df = pd.read_csv("C:/Users/yessm/Desktop/heroes_information.csv")
df.head()
#remplacer le weight par la moyenne de toute la colonne 
df['Weight'].fillna(df['Weight'].mean(), inplace=True)

#afficher le nombre des elements par publisher 
nb = len(df["Publisher"])
print("number of element:" , nb)
print(df["Publisher"].value_counts())
#remplacer les vides par le publisher le plus fréquent marvel comics 
df["Publisher"].fillna('Marvel Comics',inplace=True)
#vérifier si il ya encore des valeurs nulls 
df.isnull().sum()
```

```python
#variable categorial to numeric 
df.info()
df['Gender'].values
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform (df['Gender'])
df
df.info()
df['Gender'].values
```

```python
#creation d'un dictionnaire d'alignment 
clean = {
  "Alignment":{ "good":0 ,"bad" : 1 , "neutral" :2} 
 }
#remplacer les valeurs par les valeurs du dictionnaire 
df.replace (clean,inplace=True)
df
```

```python
one_hot =pd.get_dummies(df['Hair color'])
df=df.drop('Hair color',axis = 1)
df=df.join(one_hot)
df
```
